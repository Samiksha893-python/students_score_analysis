Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student-scores.csv')
df.head() # show first five rows
```

```python
df.info() # check data
```

```python
df.shape # check number of rows and columns
```

```python
df.isnull().sum() # check null value
```

```python
df.duplicated().sum() # check duplicate value
```

```python
df.columns # see columns name
```

```python
df.rename(columns = {'id':'ID','first_name':'First_name','last_name':'Last_name', 'email':'Email', 'gender':'Gender','part_time_job':'Part_time_job','absence_days':'Absence_days', 'extracurricular_activities':'Extracurricular_activities', 'weekly_self_study_hours':'Weekly_self_study_hours', 'semester':'Semester','career_aspiration':'Career_aspiration', 'math_score':'Math_score', 'history_score':'History_score', 'physics_score':'Physics_score', 'chemistry_score':'Chemistry_score', 'biology_score':'Biology_score', 'english_score':'English_score', 'geography_score':'Geography_score'}, inplace = True)
df   # replace name
```

```python
df['Name'] = df['First_name'] + ' ' + df['Last_name']
# add new colums
```

```python
df.drop(['First_name','Last_name'], axis = 1, inplace = True)
df  # remove colums
```

```python
df['Total_marks'] = df['Math_score'] + df['History_score'] + df['Physics_score'] + df['Chemistry_score'] + df['Biology_score'] + df['English_score'] + df['Geography_score']
df['Average_marks'] = df['Total_marks']/7
df
```

```python
df.columns
```

```python
columns = ['Name'] + [columns for columns in df.columns if columns != 'Name']
df = df[columns]
df  # Change
```

```python
df.sort_index(inplace=True)
df # sort
```

```python
df['Total_marks'] = df['Total_marks'].astype(float)
df # type conversion
```

```python
np.max(df['Average_marks'])
```

```python
df.groupby('Name')['Average_marks'].agg(['max', 'min'])
```

```python
top_student = df.loc[df['Average_marks'].idxmax()]
print(top_student)
# find topper students
```

```python
last_student = df.loc[df['Average_marks'].idxmin()]
print(last_student)
# last student
```

```python
plt.pie(df['Gender'].value_counts(), labels=df['Gender'].unique(), autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

In this graph, show the equality of boys and girls.

```python
plt.bar(df['Part_time_job'], df['Extracurricular_activities'])
plt.xlabel('Part_time_job')
plt.ylabel('Extracurricular_activities')
plt.title('Part_time_job vs Extracurricular_activities')
plt.show()
```

```python
plt.hist(df['Total_marks'], bins=5, width=30)
plt.xlabel('Total_marks')
plt.ylabel('Frequency')
plt.title('Distribution of Total_marks')
plt.show()
```

```python
plt.subplot(1, 2, 1) # Use 1 row, 2 columns for subplots
plt.hist(df['Math_score'], bins=6, width=5) # Create a histogram for Total_marks
plt.title('Distribution of math_score')
plt.xlabel('Math_score')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2) # Second subplot for Average_marks
plt.hist(df['Average_marks'], bins=6, width=5) # Create a histogram for Average_marks
plt.title('Distribution of Average Marks')
plt.xlabel('Average Marks')
plt.ylabel('Frequency')

plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
```
